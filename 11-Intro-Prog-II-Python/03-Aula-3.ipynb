Um gráfico 2D é um conjunto de pontos que relaciona valores de uma variável independente (normalmente x) com valores de uma variável dependente (normalmente y). 
Portanto, para formarmos um gráfico, precisamos de uma lista (ou vetor) de valores da variável independente e uma lista (ou vetor), de mesmo tamanho, dos valores correspondentes da
variável dependente, organizados de forma que os valores correspondentes estejam nas mesmas posições nos dois conjuntos de dados.

````python
# é fixado o intervalo de 0 a 4 π para a variável independente, x, este intervalo é dividido em 100 partes iguais (101 pontos) e, usando a estrutura iterativa for, para cada valor de x é calculada a função cosseno (função que representa a variável dependente,y). Os valores das variáveis independente e dependente são armazenados, respectivamente, nas listas x e y e, para montar o gráfico é utilizado a função plot(x, y) do pyplot.

from math import pi , cos

n = 101
x = [0]* n
y = [0]* n
x_min = 0.0
x_max = 4.0 * pi
delta_x = (x_max - x_min ) / float (n -1)

for i in range (0, n):
    x[i] = i * delta_x
    y[i] = cos(x[i])

plt.plot (x,y)
plt.show()
````

#### Inserção de linhas

hlines() e vlines. Para as linhas verticais deve-se fornecer o valor de x (ou lista com os valores)

#### Preencher espaço entre funções

função ```fill\_between()```
- Os argumentos desta função são as listas com os valores
da variável independente e das duas variáveis dependentes

````python
import matplotlib . pyplot as plt
from math import pi , cos , sin

n = 101
x = [0]* n
y1 = [0]* n; y2 = [0]* n
x_min = 0.0; x_max = 6.0 * pi
delta_x = ( x_max - x_min ) / float (n -1)

for i in range (0, n):
    x[i] = i * delta_x
    y1[i] = cos (x[i]); y2[i] = sin (x[i])
plt.plot (x,y1 , label =' Cosseno', c = (0.1 ,0.5 ,0.7))
plt.plot (x,y2 , label = 'Seno', c = '#0fbb52')
plt.legend ()
plt.fill_between (x,y1 ,y2 , color = (0.8 ,0.9 ,0.7))
````

#### Subplot

````python
import matplotlib.pyplot as plt
from math import pi, cos, sin

n = 1001
x = [0]* n
y1 = [0]* n; 
y2 = [0]* n; 
y3 = [0]* n; 
y4 = [0]* n; 
y5 = [0]* n; 
y6 = [0]* n
x_min = 0.0; x_max = 6.0 * pi
delta_x = ( x_max - x_min ) / float (n -1)

for i in range (0, n):
    x[i] = i * delta_x
    y1[i] = cos (x[i]); y2[i] = cos (2 * x[i]); y3[i] = cos (3 * x[i])
    y4[i] = sin (x[i]); y5[i] = sin (2 * x[i]); y6[i] = sin (3 * x[i])
plt.subplot(3 ,2 ,1)
plt.plot(x,y1)
plt.title(r'$\cos\ left (x\ right )$ ')
plt.subplot(3 ,2 ,3)
plt.plot(x,y2)
plt.title(r'$\cos\ left (2 \, x\ right )$ ')
plt.subplot(3 ,2 ,5)
plt.plot(x,y3)
plt.title(r'$\cos\ left (3 \, x\ right )$ ')
plt.subplot(3 ,2 ,2)
plt.plot(x,y4)
plt.title(r'$\sin\ left (x\ right )$ ')
plt.subplot(3 ,2 ,4)
plt.plot(x,y5)
plt.title(r'$\sin\ left (2 \, x\ right )$ ')
plt.subplot(3 ,2 ,6)
plt.plot(x,y6)
plt.title(r'$\sin\ left (3 \, x\ right )$ ')
plt.tight_layout()
plt.show()
````

#### dar destaque

A função ```axes([left, bottom, width, height])``` pode ser usada para dar destaque a uma parte específica do gráfico sendo apresentado.

````python
import matplotlib.pyplot as plt
from math import pi, sin

n = 101
x = [0]* n; y = [0]* n
x_min = 0.0; x_max = 6.0 * pi
delta_x = ( x_max - x_min ) / float (n -1)
for i in range (0, n):
    x[i] = i * delta_x
    y[i] = sin(x[i])
x2 = [0]* n; y2 = [0]* n
x2_min = 0.0; x2_max = 1.0 * pi
delta_x2 = ( x2_max - x2_min ) / float (n -1)
for i in range (0, n):
    x2[i] = i * delta_x2
    y2[i] = sin (x2[i])
plt.plot(x,y)
plt.xlabel('x ')
plt.ylabel('y ')
plt.axes([0.2 ,0.2 ,0.25 ,0.25])
plt.plot(x2 ,y2)
plt.xlabel('x2 ')
plt.ylabel('y2 ')
plt.show()
````

## 9.2 Introdução aos gráficos 3D

a representação de cada ponto necessita de um trio de valores, sendo dois para as variáveis independentes (x e y) e um
para a variável dependente (z). 
o script a seguir gera um gráfico 3D que apresenta um único ponto com x = 3, y = 5 e z = 7.

````python
import matplotlib . pyplot as plt

ax = plt.axes( projection = "3d")
ax.scatter(3 ,5 ,7)
plt.show()
````

é necessário criar um objeto (nomeado ax no script) que receberá todas as informações do gráficos, isso é feito usando o método axes como o argumento projection = ‘‘3d’’.

No Script 9.22 é apresentada a forma de montar um gráfico com vários pontos, sendo que para gerar os 200 pontos, com valores entre zero e cem em cada uma das coordenadas, utilizou-se o método ```random.randint``` do pacote numpy.

````python
import numpy as np
import matplotlib.pyplot as plt

x_data = np.random.randint(0, 100 , (200 ,))
y_data = np.random.randint(0, 100 , (200 ,))
z_data = np.random.randint (0, 100 , (200 ,))

ax = plt.axes(projection = "3d")
ax.scatter(x_data , y_data , z_data , marker = "v")
plt.show()
````
