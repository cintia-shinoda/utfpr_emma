## 9.1 Biblioteca pyplot do matplotlib

````python
import matplotlib.pyplot as plt
````

Um gráfico 2D é um conjunto de pontos que relaciona valores de uma variável independente (normalmente x) com valores de uma variável dependente (normalmente y). 
Portanto, para formarmos um gráfico, precisamos de uma lista (ou vetor) de valores da variável independente e uma lista (ou vetor), de mesmo tamanho, dos valores correspondentes da
variável dependente, organizados de forma que os valores correspondentes estejam nas mesmas posições nos dois conjuntos de dados.

````python
# é fixado o intervalo de 0 a 4 π para a variável independente, x, este intervalo é dividido em 100 partes iguais (101 pontos) e, usando a estrutura iterativa for, para cada valor de x é calculada a função cosseno (função que representa a variável dependente,y). Os valores das variáveis independente e dependente são armazenados, respectivamente, nas listas x e y e, para montar o gráfico é utilizado a função plot(x, y) do pyplot.

from math import pi , cos

n = 101
x = [0]* n
y = [0]* n
x_min = 0.0
x_max = 4.0 * pi
delta_x = (x_max - x_min ) / float (n -1)

for i in range (0, n):
    x[i] = i * delta_x
    y[i] = cos(x[i])

plt.plot (x,y)
plt.show()
````

````python
# salvar a figura gerada usando a função savefig(‘nomefig’).
# é gerado um gráfico para função cosseno no intervalo de 0 a 6 π, sendo esta figura salva no formato pdf. Por padrão, a figura é salva na mesma pasta em que o script do python está contido.

from math import pi , cos

n = 101
x = [0] * n
y = [0] * n
x_min = 0.0
x_max = 6.0 * pi
delta_x = ( x_max - x_min ) / float (n -1)

for i in range (0, n):
    x[i] = i * delta_x
    y[i] = cos(x[i])
plt.plot (x,y)
plt.savefig ('Fig-02.pdf')
plt.show ()
````

````python
# Geração do gráfico das funções cosseno e seno entre 0 e 6 π

from math import pi , cos , sin

n= 101
x = [0]* n
y1 = [0]* n
y2 = [0]* n
x_min = 0.0
x_max = 6.0 * pi
delta_x = (x_max - x_min) / float (n -1)

for i in range (0, n):
    x[i] = i * delta_x
    y1[i] = cos (x[i])
    y2[i] = sin (x[i])

plt.plot(x,y1)
plt.plot(x,y2)

plt.show ()
````

````python
# 

from math import pi , cos , sin

n= 101
x = [0]* n
y1 = [0]* n
y2 = [0]* n
x_min = 0.0
x_max = 6.0 * pi
delta_x = (x_max - x_min) / float (n -1)

for i in range (0, n):
    x[i] = i * delta_x
    y1[i] = cos (x[i])
    y2[i] = sin (x[i])

plt.plot(x,y1, label='Cosseno')
plt.plot(x,y2, label = 'Seno')
plt.legend()
plt.show ()
````

### Cores

usar argumento color (ou simplesmente c) em cada função plot(). 
- ‘c’ → Cyan;
- ‘m’ → Magenta;
- ‘w’ → White;
- ‘b’ → Blue;
- ‘g’ → Green;
- ‘r’ → Red;
- ‘k’ → Black;
- ‘y’ → Yellow.
<br>

caso o usuário queira configurar corres diferentes destas, é possível fornecer as proporções RGB (Red, Green e Blue) para formar a cor desejada, sendo que há duas formas diferentes de fornecer estas proporções, 
- por meio de uma tupla contendo três valores entre zero e um (por exemplo, c = (0.7, 0.2, 0.8) ), 
- três conjuntos de dois algarismos hexadecimais (que vão de
00 a ff) precedidos de # (tudo isso entre aspas, por exemplo, c =‘#1f4cbb’).
